# I'm making columns here to show the percentage change across time. I have four quartiles and one column for the net change.

```
life_expectancy["percentage_change_62_74"] = ((life_expectancy["1974"] - life_expectancy["1962"]) / life_expectancy["1962"]) * 100
```

```
life_expectancy["percentage_change_75_87"] = ((life_expectancy["1987"] - life_expectancy["1975"]) / life_expectancy["1975"]) * 100
```

```
life_expectancy["percentage_change_88_01"] = ((life_expectancy["2001"] - life_expectancy["1988"]) / life_expectancy["1988"]) * 100
```

```
life_expectancy["percentage_change_89_13"] = ((life_expectancy["2013"] - life_expectancy["1989"]) / life_expectancy["1989"]) * 100
```

```
life_expectancy["net_percentage_change_62_13"] = ((life_expectancy["2013"] - life_expectancy["1962"]) / life_expectancy["1962"]) * 100
```
